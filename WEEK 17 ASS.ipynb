Q1. What is the Probability density function?

ANS- 
A probability density function (PDF) is a statistical function that describes the likelihood of a continuous random variable taking on a particular value. Unlike a probability mass function (PMF), which is used for discrete variables, a PDF provides the density of probabilities over an interval.

Q2. What are the types of Probability distribution?

Probability distributions can be broadly categorized into two main types: **discrete probability distributions** and **continuous probability distributions**. Here’s a breakdown of each type along with some common examples:

### 1. Discrete Probability Distributions
These distributions apply to discrete random variables, which can take on a countable number of values. Key characteristics include specific probabilities assigned to each outcome.

**Common Examples:**
- **Binomial Distribution**: Models the number of successes in a fixed number of independent Bernoulli trials.
- **Poisson Distribution**: Models the number of events occurring within a fixed interval of time or space, given a constant mean rate.
- **Geometric Distribution**: Models the number of trials needed for the first success in a series of independent Bernoulli trials.
- **Hypergeometric Distribution**: Models successes in a sequence without replacement, typically used in sampling scenarios.

### 2. Continuous Probability Distributions
These distributions apply to continuous random variables, which can take on an infinite number of values within a given range. Instead of probabilities for specific outcomes, they provide probabilities for intervals.

**Common Examples:**
- **Normal Distribution**: Bell-shaped curve characterized by its mean and standard deviation, widely used in statistics.
- **Exponential Distribution**: Models the time between events in a Poisson process, often used in survival analysis and reliability engineering.
- **Uniform Distribution**: All outcomes are equally likely within a certain interval.
- **Gamma Distribution**: Generalizes the exponential distribution; useful in queuing models and reliability analysis.

### 3. Other Notable Distributions
- **Log-Normal Distribution**: Used for modeling variables that are positively skewed, such as stock prices.
- **Beta Distribution**: Defined on the interval [0, 1], often used in Bayesian statistics.
- **Chi-Squared Distribution**: Commonly used in hypothesis testing and confidence interval estimation.

Each distribution has specific properties and applications, making them suitable for different types of data and analysis scenarios.

Q3. Write a Python function to calculate the probability density function of a normal distribution with
given mean and standard deviation at a given point.

```python
import math

def normal_pdf(x, mean, std_dev):
    """Calculate the probability density function of a normal distribution.
    
    Args:
    x (float): The point at which to evaluate the PDF.
    mean (float): The mean (μ) of the distribution.
    std_dev (float): The standard deviation (σ) of the distribution.

    Returns:
    float: The value of the PDF at point x.
    """
    coefficient = 1 / (std_dev * math.sqrt(2 * math.pi))
    exponent = -((x - mean) ** 2) / (2 * std_dev ** 2)
    return coefficient * math.exp(exponent)

# Example usage:
mean = 0
std_dev = 1
point = 1
pdf_value = normal_pdf(point, mean, std_dev)
print(f"The PDF value at x = {point} is: {pdf_value}")
```

Q4. What are the properties of Binomial distribution? Give two examples of events where binomial
distribution can be applied.

The binomial distribution has several key properties:

1. **Fixed Number of Trials**: The process consists of a fixed number of trials (n).

2. **Two Outcomes**: Each trial results in one of two outcomes: success (often coded as 1) or failure (coded as 0).

3. **Constant Probability**: The probability of success (p) remains constant for each trial.

4. **Independent Trials**: The trials are independent; the outcome of one trial does not affect the others.

5. **Discrete Distribution**: The binomial distribution is discrete, meaning it counts the number of successes in n trials.

### Examples of Binomial Distribution Events

1. **Coin Tossing**: When tossing a fair coin 10 times, the number of times it lands on heads can be modeled with a binomial distribution. Here, each toss is a trial, heads is a success, and the probability of heads (success) is 0.5.

2. **Quality Control**: In a factory, if an inspector checks 20 light bulbs and determines whether each bulb is defective (failure) or non-defective (success), the number of non-defective bulbs can be modeled as a binomial distribution. If the probability of a bulb being non-defective is 0.9, this scenario fits the binomial framework. 

These examples illustrate how the binomial distribution can be applied in various contexts involving repeated independent trials with two possible outcomes.

Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4
and plot a histogram of the results using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 1000  # Number of trials
p = 0.4   # Probability of success

# Generate random sample from a binomial distribution
sample = np.random.binomial(n, p, size=1000)

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(sample, bins=np.arange(0, n+1) - 0.5, density=True, alpha=0.7, color='blue')
plt.title('Histogram of Binomial Distribution (n=1000, p=0.4)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability Density')
plt.xticks(range(0, n+1, 50))
plt.grid()
plt.show()
```

Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution
with given mean at a given point.

To calculate the cumulative distribution function (CDF) of a Poisson distribution in Python, you can use the scipy.stats module, which provides a convenient way to work with probability distributions. Here's a function that calculates the CDF for a given mean and point:

```python
import scipy.stats as stats

def poisson_cdf(mean, x):
    """
    Calculate the cumulative distribution function (CDF) of a Poisson distribution.
    
    Parameters:
    mean (float): The mean (lambda) of the Poisson distribution.
    x (int): The point at which to evaluate the CDF.
    
    Returns:
    float: The CDF value at point x.
    """
    return stats.poisson.cdf(x, mean)

# Example usage
mean = 3.5
x = 5
cdf_value = poisson_cdf(mean, x)
print(f"The CDF of a Poisson distribution with mean {mean} at point {x} is {cdf_value:.4f}")
```

Q7. How Binomial distribution different from Poisson distribution?

The Binomial and Poisson distributions are both discrete probability distributions, but they differ in several key aspects:

### 1. **Nature of Trials**
- **Binomial Distribution**: Used for a fixed number of independent trials, each with the same probability of success. It counts the number of successes in these trials.
- **Poisson Distribution**: Used for counting the number of events occurring in a fixed interval of time or space, given that these events happen with a known constant mean rate and independently of the time since the last event.

### 2. **Parameters**
- **Binomial Distribution**: Defined by two parameters:
  - \( n \): Number of trials.
  - \( p \): Probability of success in each trial.
- **Poisson Distribution**: Defined by one parameter:
  - \( \lambda \): Average rate of occurrence (mean number of events in a given interval).

### 3. **Shape**
- **Binomial Distribution**: The shape can vary and can be symmetric or skewed, depending on the values of \( n \) and \( p \).
- **Poisson Distribution**: The shape is typically right-skewed, especially for smaller values of \( \lambda \). As \( \lambda \) increases, it can become more symmetric.

### 4. **Usage**
- **Binomial Distribution**: Appropriate for scenarios like flipping a coin a certain number of times or testing a product for defects.
- **Poisson Distribution**: Suitable for events like counting the number of emails received in an hour or the number of cars passing through a toll booth in a day.

### 5. **Relationship**
- As the number of trials \( n \) in a binomial distribution becomes large and the probability of success \( p \) becomes small such that \( np = \lambda \) remains constant, the binomial distribution can be approximated by a Poisson distribution.

### Summary Table

| Feature                | Binomial Distribution              | Poisson Distribution            |
|------------------------|------------------------------------|---------------------------------|
| Trials                 | Fixed number \( n \)              | No fixed number (rate \( \lambda \)) |
| Success Probability     | Constant \( p \)                  | Not applicable                  |
| Parameters             | \( n, p \)                        | \( \lambda \)                   |
| Shape                  | Can be symmetric or skewed        | Typically right-skewed          |
| Example Usage          | Coin toss, product testing         | Emails received, events in a time interval |

These distinctions help determine which distribution to use based on the context of the problem at hand.

Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the
sample mean and variance.

To generate a random sample from a Poisson distribution with a mean of 5, and then calculate the sample mean and variance using Python. Here’s how you can do that:

```python
import numpy as np

# Set the parameters
mean = 5
sample_size = 1000

# Generate random sample from a Poisson distribution
sample = np.random.poisson(mean, sample_size)

# Calculate sample mean and variance
sample_mean = np.mean(sample)
sample_variance = np.var(sample)

# Print the results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Variance: {sample_variance:.2f}")
```

Q9. How mean and variance are related in Binomial distribution and Poisson distribution?

In both the Binomial and Poisson distributions, the mean and variance have specific relationships that are fundamental to their properties.

### Binomial Distribution

For a Binomial distribution with parameters:
- \( n \): number of trials
- \( p \): probability of success

The mean (\( \mu \)) and variance (\( \sigma^2 \)) are given by:

- **Mean**: 
  \[
  \mu = np
  \]

- **Variance**: 
  \[
  \sigma^2 = np(1 - p)
  \]

### Relationship in Binomial Distribution
- The mean and variance are both functions of \( n \) and \( p \).
- As \( p \) approaches 0 or 1 (meaning fewer successes), the variance decreases relative to the mean.

### Poisson Distribution

For a Poisson distribution with parameter:
- \( \lambda \): average rate of occurrence (mean)

The mean and variance are both given by:

- **Mean**: 
  \[
  \mu = \lambda
  \]

- **Variance**: 
  \[
  \sigma^2 = \lambda
  \]

### Relationship in Poisson Distribution
- The mean and variance are equal (\( \mu = \sigma^2 \)).
- This property is unique to the Poisson distribution and reflects its behavior in counting events over a fixed interval.

### Summary of Relationships

| Distribution      | Mean                  | Variance              | Relationship             |
|-------------------|-----------------------|-----------------------|--------------------------|
| Binomial          | \( \mu = np \)       | \( \sigma^2 = np(1 - p) \) | Mean depends on \( n \) and \( p \) |
| Poisson           | \( \mu = \lambda \)  | \( \sigma^2 = \lambda \) | Mean equals variance     |

In summary, while both distributions relate their mean and variance to their respective parameters, the Binomial distribution's variance depends on both \( n \) and \( p \), while the Poisson distribution has the property that its mean equals its variance.

Q10. In normal distribution with respect to mean position, where does the least frequent data appear?

In a normal distribution, the data is symmetrically distributed around the mean. The least frequent data points appear in the tails of the distribution.

### Key Points:

1. **Mean Position**: The mean (denoted as \(\mu\)) is the center of the distribution. It divides the distribution into two equal halves.

2. **Tails of the Distribution**: The tails extend infinitely in both directions (left and right) from the mean. As you move away from the mean, the frequency of data points decreases.

3. **Least Frequent Data**: The least frequent data appears at the extremes—farther from the mean, in both the left tail (for lower values) and the right tail (for higher values). The probabilities of finding data points in these regions are very low compared to those near the mean.

### Visualization
- **Peak**: The highest point of the normal distribution curve is at the mean, where the most frequent data points are found.
- **Tails**: The areas at the far left and right of the curve represent the least frequent occurrences, often corresponding to values that are several standard deviations away from the mean.
